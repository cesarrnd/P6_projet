```python
import os
import numpy as np
import h5py
import cv2
import yaml
from tqdm import tqdm

# ── Chemins ──
TRAIN_TD   = "/content/drive/My Drive/data_p6/moorea_2019-02-19_004_td_610500000_670500000_td.h5"
TRAIN_BBOX = "/content/drive/My Drive/data_p6/moorea_2019-02-19_004_td_610500000_670500000_bbox.npy"
VAL_TD     = "/content/drive/My Drive/data_p6/test.h5"
VAL_BBOX   = "/content/drive/My Drive/data_p6/test.npy"
OUT_DIR    = "/content/drive/My Drive/data_p6/data_gen1_coco"
T          = 4
OUT_H, OUT_W = 320, 320

def timesurface_to_rgb(ts_frame):
    H, W = ts_frame.shape[1], ts_frame.shape[2]
    img = np.full((H, W, 3), 127, dtype=np.uint8)
    pos, neg = ts_frame[0], ts_frame[1]
    def norm(x):
        mx = x.max()
        return np.zeros_like(x, dtype=np.uint8) if mx == 0 else (x.astype(np.float32)/mx*255).astype(np.uint8)
    pos_n, neg_n = norm(pos), norm(neg)
    img[pos > 0, 0] = pos_n[pos > 0]
    img[pos > 0, 1] = 0
    img[pos > 0, 2] = 0
    img[neg > 0, 0] = 0
    img[neg > 0, 1] = 0
    img[neg > 0, 2] = neg_n[neg > 0]
    return img

def process_split(events_h5, bbox_npy, split_name, name):
    out_split = os.path.join(OUT_DIR, split_name)
    os.makedirs(out_split, exist_ok=True)
    txt_path = os.path.join(OUT_DIR, f"{split_name}.txt")

    bbox = np.load(bbox_npy, allow_pickle=True)
    if "ts" in bbox.dtype.names:
        bbox.dtype.names = ["t" if n == "ts" else n for n in bbox.dtype.names]

    unique_ts, first_idx = np.unique(bbox["t"], return_index=True)
    groups = np.split(bbox, first_idx[1:])
    ts_min, ts_max = bbox["t"].min(), bbox["t"].max()
    print(f"\n{split_name}: {len(groups)} timestamps bbox")

    with h5py.File(events_h5, "r") as f:
        N = f["events"].shape[0]
        H_orig = f["events"].shape[2]
        W_orig = f["events"].shape[3]
        print(f"events shape: {f['events'].shape}")

        written = 0
        with open(txt_path, "w") as list_f:
            for idx, group in enumerate(tqdm(groups, desc=split_name)):
                ts_val = group["t"][0]
                fc = int((ts_val - ts_min) / (ts_max - ts_min) * (N - 1)) if ts_max != ts_min else N // 2
                start = max(0, min(fc - T//2, N - T))
                end   = start + T

                frames = f["events"][start:end]  # (T, 2, H, W)
                if frames.shape[0] < T:
                    pad = np.zeros((T - frames.shape[0], 2, H_orig, W_orig), dtype=frames.dtype)
                    frames = np.concatenate([pad, frames], axis=0)

                img_out = np.full((T, OUT_H, OUT_W, 3), 127, dtype=np.uint8)
                for t in range(T):
                    img_out[t] = cv2.resize(timesurface_to_rgb(frames[t]), (OUT_W, OUT_H))

                labels = []
                for box in group:
                    cx = np.clip((float(box["x"]) + float(box["w"])/2) / W_orig, 0, 1)
                    cy = np.clip((float(box["y"]) + float(box["h"])/2) / H_orig, 0, 1)
                    nw = np.clip(float(box["w"]) / W_orig, 0, 1)
                    nh = np.clip(float(box["h"]) / H_orig, 0, 1)
                    if nw > 0 and nh > 0:
                        labels.append([int(box["class_id"]), cx, cy, nw, nh])

                if not labels:
                    continue

                img_path = os.path.join(out_split, f"img_{name}_{idx:06d}.npy")
                lbl_path = os.path.join(out_split, f"img_{name}_{idx:06d}.txt")
                np.save(img_path, img_out)
                np.savetxt(lbl_path, np.array(labels, dtype=np.float32), fmt="%.6f")
                list_f.write(img_path + "\n")
                written += 1

        print(f"→ {written} samples écrits")

os.makedirs(OUT_DIR, exist_ok=True)
process_split(TRAIN_TD, TRAIN_BBOX, "train", "moorea_train")
process_split(VAL_TD,   VAL_BBOX,   "val",   "moorea_val")

# ── Générer YAML ──
bbox_train = np.load(TRAIN_BBOX, allow_pickle=True)
unique_cls  = np.unique(bbox_train["class_id"])
nc = int(unique_cls.max()) + 1
names = {int(c): f"class_{int(c)}" for c in unique_cls}

yaml_path = os.path.join(OUT_DIR, "gen1_moorea.yaml")
with open(yaml_path, "w") as f:
    yaml.dump({"path": OUT_DIR, "train": "train.txt", "val": "val.txt",
               "nc": nc, "names": names}, f, default_flow_style=False)

print(f"\n✅ YAML généré : {yaml_path}")
print(f"   nc={nc}, classes={names}")
```

```python
import yaml
import numpy as np

TRAIN_BBOX = "/content/drive/My Drive/data_p6/moorea_2019-02-19_004_td_610500000_670500000_bbox.npy"
yaml_path  = "/content/drive/My Drive/data_p6/data_gen1_coco/gen1_moorea.yaml"

# Voir les classes réelles
bbox = np.load(TRAIN_BBOX, allow_pickle=True)
unique_cls = sorted(np.unique(bbox['class_id']).tolist())
print("Classes originales:", unique_cls)

# Remapping : classe originale → index 0-based
remap = {int(c): i for i, c in enumerate(unique_cls)}
print("Remapping:", remap)

# Noms Prophesee Gen4 (d'après ton gen4.yaml original)
prophesee_names = {
    0: 'pedestrian', 1: 'two_wheeler', 2: 'car',
    3: 'truck', 4: 'bus', 5: 'traffic_sign', 6: 'traffic_light'
}
names = {i: prophesee_names.get(int(c), f'class_{int(c)}') for i, c in enumerate(unique_cls)}
print("Names finaux:", names)

# Réécrire le yaml
with open(yaml_path, 'w') as f:
    yaml.dump({
        'path':  "/content/drive/My Drive/data_p6/data_gen1_coco",
        'train': 'train.txt',
        'val':   'val.txt',
        'nc':    len(unique_cls),
        'names': names,
    }, f, default_flow_style=False)

print(f"✅ YAML corrigé : nc={len(unique_cls)}")
```

```python
import os
from tqdm import tqdm

OUT_DIR = "/content/drive/My Drive/data_p6/data_gen1_coco"

for split in ['train', 'val']:
    folder = os.path.join(OUT_DIR, split)
    txts = [f for f in os.listdir(folder) if f.endswith('.txt')]
    for fname in tqdm(txts, desc=f"Remapping {split}"):
        fpath = os.path.join(folder, fname)
        rows = np.loadtxt(fpath)
        if rows.ndim == 1:
            rows = rows[None, :]
        if len(rows) == 0:
            continue
        # Remappe la colonne 0 (class_id)
        rows[:, 0] = np.array([remap[int(c)] for c in rows[:, 0]])
        np.savetxt(fpath, rows, fmt='%.6f')

print("✅ Labels remappés")
```

```python
import numpy as np
import yaml

TRAIN_BBOX = "/content/drive/My Drive/data_p6/moorea_2019-02-19_004_td_610500000_670500000_bbox.npy"
VAL_BBOX   = "/content/drive/My Drive/data_p6/test.npy"
yaml_path  = "/content/drive/My Drive/data_p6/data_gen1_coco/gen1_moorea.yaml"

# Collecter toutes les classes des deux splits
bbox_train = np.load(TRAIN_BBOX, allow_pickle=True)
bbox_val   = np.load(VAL_BBOX,   allow_pickle=True)

all_cls = sorted(set(
    np.unique(bbox_train['class_id']).tolist() +
    np.unique(bbox_val['class_id']).tolist()
))
print("Toutes les classes:", all_cls)

remap = {int(c): i for i, c in enumerate(all_cls)}
print("Remapping:", remap)

prophesee_names = {
    0: 'pedestrian', 1: 'two_wheeler', 2: 'car',
    3: 'truck', 4: 'bus', 5: 'traffic_sign', 6: 'traffic_light'
}
names = {i: prophesee_names.get(int(c), f'class_{int(c)}') for i, c in enumerate(all_cls)}
print("Names:", names)

# Réécrire le yaml
with open(yaml_path, 'w') as f:
    yaml.dump({
        'path':  "/content/drive/My Drive/data_p6/data_gen1_coco",
        'train': 'train.txt',
        'val':   'val.txt',
        'nc':    len(all_cls),
        'names': names,
    }, f, default_flow_style=False)

print(f"✅ YAML corrigé : nc={len(all_cls)}")
```

```python
import os
from tqdm import tqdm

OUT_DIR = "/content/drive/My Drive/data_p6/data_gen1_coco"

for split in ['train', 'val']:
    folder = os.path.join(OUT_DIR, split)
    txts = [f for f in os.listdir(folder) if f.endswith('.txt')]
    for fname in tqdm(txts, desc=f"Remapping {split}"):
        fpath = os.path.join(folder, fname)
        try:
            rows = np.loadtxt(fpath)
        except Exception:
            continue
        if rows.size == 0:
            continue
        if rows.ndim == 1:
            rows = rows[None, :]
        # Ignorer les classes inconnues (ne devrait plus arriver)
        valid = np.array([int(c) in remap for c in rows[:, 0]])
        rows = rows[valid]
        if len(rows) == 0:
            os.remove(fpath)
            continue
        rows[:, 0] = np.array([remap[int(c)] for c in rows[:, 0]])
        np.savetxt(fpath, rows, fmt='%.6f')

print("✅ Labels remappés")
```

```python
with open('/content/SpikeYOLO/SpikeYOLO_for_Gen1/train.py', 'w') as f:
    f.write('''import os
import shutil
from ultralytics import YOLO

os.environ["WANDB_DISABLED"] = "true"
os.environ["WANDB_MODE"] = "disabled"

try:
    import wandb
    wandb.init(mode="disabled")
except:
    pass

SAVE_DIR = "/content/drive/My Drive/data_p6/checkpoints"
os.makedirs(SAVE_DIR, exist_ok=True)

model = YOLO("snn_yolov8s.yaml")

def save_callback(trainer):
    for f in ["best.pt", "last.pt"]:
        src = os.path.join(str(trainer.save_dir), "weights", f)
        if os.path.exists(src):
            shutil.copy(src, os.path.join(SAVE_DIR, f))
            print(f"✅ {f} sauvegardé sur Drive")

model.add_callback("on_train_epoch_end", save_callback)

model.train(
    data="/content/drive/My Drive/data_p6/data_gen1_coco/gen1_moorea.yaml",
    device=0,
    epochs=50,
    batch=8,
    imgsz=320,
)
''')
print("✅ train.py écrit")
```
